```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/mentalh_health_dataset_bersih.csv")

df
```

###Distribusi Pekerjaan

```python
plt.figure(figsize=(10, 8))
plt.bar(df['Occupation'].value_counts().index, df['Occupation'].value_counts().values, color='skyblue')
plt.title('Distribusi Pekerjaan')
plt.xlabel('Pekerjaan')
plt.ylabel('Jumlah')
plt.xticks(rotation=30)
plt.grid()
plt.show()
```

Insight:
dari bar chart diatas, kategori pekerjaan "Other" memiliki jumlah tertinggi dibandingkan kategori pekerjaan lainnya, seperti Healthcare, Engineering, Finance, Sales, Education, dan IT, yang memiliki jumlah responden relatif seimbang.

tindakan:
melakukan analisis lanjutan untuk mengetahui lebih lanjut terkait pekerjaan pada kategori "other" untuk dapat mengambil keputusan yang lebih tepat lagi dan agar dapat lebih memahami faktor risiko yang memengaruhi kesehatan mental khususnya faktor pekerjaan.

##Distribusi Gender

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Gender'].value_counts(), labels=df['Gender'].value_counts().index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold'])
plt.title('Distribusi Gender')
plt.show()
```

Insight:
Distribusi gender menunjukkan variasi yang cukup merata menunjukkan bahwa ada keragaman identitas gender dalam survei yang dilakukan. dan pada kategori "Prefer not to say" mungkin karena adanya isu privasi dan ketidaknyaman untuk mengungkapkan identitas gender mereka.

Tindakan:
Melakukan tinjauan kembali untuk memastikan kelompok gender terwakili. dan melakukan evaluasi terkait kenapa banyak yang tidak ingin mengungkapkan identitas gender mereka, dengan cara menawarkan opsi yang lebih aman dan anonim.

##Distribusi Jam Kerja

```python
plt.figure(figsize=(7, 5))
sns.histplot(df['Work_Hours'], bins=20, kde=True, color='blue', edgecolor='black')
plt.title('Distribusi Jam Kerja')
plt.xlabel('Durasi (Jam)')
plt.ylabel('Jumlah')
plt.show()
```

Insight:
Data menunjukkan distribusi frekuensi jam kerja terlihat cukup merata di berbagai rentang jam kerja, dan terdapat puncak frekuensi di 65-70 jam yang berarti ada sejumlah individu yang bekerjja dalam rentang waktu tersebut.

Tindakan:
Melakukan analisis lebih lanjut untuk mengetahui faktor-faktor apa saja khususnya pada rentang 65-70 jam alasan mereka bekerja di rentang waktu tersebut. Selain itu, organisai dapat menyesuaikan kebijakan terkait jam kerja untuk meningkatkan keseimbangan dan kesehatan mental pekerjanya.

##Perbandingan Antara Durasi Jam kerja dan Durasi Jam Tidur

```python
df['Work_Hours_aday'] = df['Work_Hours'] / 7
plt.figure(figsize=(10, 8))
plt.scatter(df['Work_Hours_aday'], df['Sleep_Hours'], color='blue', alpha=0.5)
plt.title('Durasi Jam Kerja Perhari vs Durasi Jam Tidur Perhari')
plt.xlabel('Jam Kerja Perhari')
plt.ylabel('Durasi (menit)')
plt.grid()
plt.show()
```

Insight:
Berdasarkan Scatter plot bahwa data tersebar secara acak yang berarti tidak ada korelasi antara jam tidur perhari dan jam kerja perhari. Jadi, jam kerja yang tinggi tidak selalu memiliki jam tidur yang lebih sedikit atau sebaliknya. rentang jam kerja dan jam tidur cukup bervariasi.

Tindakan:
Melakukan Analisis lebih lanjut untuk mengetahui lebih lanjut perbandingan antara jam kerja dan jam tidur dan faktor apa saja yang membuat antara jam kerja dan jam tidur sangat bervariasi.
